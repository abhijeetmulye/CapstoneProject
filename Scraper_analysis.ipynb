```python
# TODO add image and put this code into an appendix at the bottom
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

X = sf_US[['goal', 'is_CA']]
y = sf_US['pledged']

## fit a OLS model with intercept on TV and Radio
X = sm.add_constant(X)
est = sm.OLS(y, X).fit()

## Create the 3d plot -- skip reading this
# TV/Radio grid for 3d plot
xx1, xx2 = np.meshgrid(np.linspace(X.goal.min(), X.goal.max(), 100), 
                       np.linspace(X.is_CA.min(), X.is_CA.max(), 100))
# plot the hyperplane by evaluating the parameters on the grid
Z = est.params[0] + est.params[1] * xx1 + est.params[2] * xx2

# create matplotlib 3d axes
fig = plt.figure(figsize=(12, 8))
ax = Axes3D(fig, azim=-115, elev=15)

# plot hyperplane
surf = ax.plot_surface(xx1, xx2, Z, cmap=plt.cm.RdBu_r, alpha=0.6, linewidth=0)

# plot data points - points over the HP are white, points below are black
resid = y - est.predict(X)
ax.scatter(X[resid >= 0].goal, X[resid >= 0].is_CA, y[resid >= 0], color='black', alpha=1.0, facecolor='white')
ax.scatter(X[resid < 0].goal, X[resid < 0].is_CA, y[resid < 0], color='black', alpha=1.0)

# set axis labels
ax.set_xlabel('Goal')
ax.set_ylabel('California')
ax.set_zlabel('Pledged')
```
